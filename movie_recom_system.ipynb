```python
import ast

def convert3(obj):
    try:
        # Safely parse the string representation of the list
        cast_list = ast.literal_eval(obj)
        # Extract the 'name' from the first 3 elements using a list comprehension
        return [actor['name'] for actor in cast_list[:3]]
    except (ValueError, SyntaxError):
        # Return an empty list if a row contains malformed data or NaN
        return []

# Apply the function to overwrite the cast column with the structured names
main_data['cast'] = main_data['cast'].apply(convert3)

# Display the updated DataFrame to verify the results
display(main_data.head())
```

```python
import ast

def fetch_director(obj):
    try:
        # Safely parse the string representation of the crew list
        crew_list = ast.literal_eval(obj)

        # Look for the person whose job is 'Director'
        for member in crew_list:
            if member['job'] == 'Director':
                return member['name']

        # Return None or 'Unknown' if no director is listed in the crew
        return None
    except (ValueError, SyntaxError):
        # Handle empty/missing (NaN) values safely
        return None

# Create a new column 'director' or overwrite 'crew'
main_data['crew'] = main_data['crew'].apply(fetch_director)

# Verify the changes
display(main_data['crew'].head())
```

```python
def stem(text):
  y = []

  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

```python
def recommend(movie):
  movie_index = new_df[new_df['title_y'] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

  for i in movies_list:
    print(new_df.iloc[i[0]].title_y)
```
